```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from keras import Input
from keras.layers import Dense

from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv(r"C:\Users\deanu\OneDrive\Desktop\Data Sets\data.csv")
data.head()
```

```python
#Print columns
print(data.columns)
```

```python
#Create target column
# M = 0    B = 1

if 'diagnosis' in data.columns:
    data['target'] = data['diagnosis'].map({'M':1, 'B':0})
else:
    raise ValueError("Not found")

data = data.drop(columns=['diagnosis'])
```

```python
# Drop columns ---> Unnamed: 32  , id
data = data.drop(columns=['id', 'Unnamed: 32'])
data.head()
```

```python
# Separating the features and target
X = data.drop(columns=['target'])
y = data['target']
```

```python
X.head()
```

```python
y.head()
```

```python
# SPlitting the aata into training data and testing data
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
model = keras.Sequential([
    Input(shape = (30,)),
    Dense(60,activation='relu'),
    Dense(30,activation='relu'),
    Dense(1,activation='sigmoid')
])

from keras.optimizers import Adam
from keras.losses import BinaryCrossentropy
model.compile(
    optimizer = Adam(learning_rate = 1e-3),
    loss = BinaryCrossentropy(),
    metrics = ['accuracy']
)

model.fit(X_train,y_train,epochs=100)
```

```python
# Accuracy on trainning data
model.evaluate(X_train,y_train)
```

```python
# Accuracy on testing data
model.evaluate(X_test,y_test)
```
